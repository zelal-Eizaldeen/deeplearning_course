The binary classification problems has two output values.

A common example of **a binary classification method** is if we want to do a medical diagnosis. We want to predict if a patient has a specific condition based on a number of input variables.

Another common problem that we haven't seen so far is **a regression problem**. Here we try to **predict a numerical value rather than a class**.

We will look at how we can choose **different networks** to solve **different types of problem**s. The ***type of problem that we wanna solve will affect the type of the output unit and its loss function***.

# Multiclass Classification

Moving on to multi-class classification. Here we want the output layer to be able to have **one output for each class**. To do that, we use **the softmax activation function.** **Each of the outputs will have a range between zero and one, but with an additional restriction that the sum** **of all the outputs will total 1.0. **That means that we can then interpret these outputs as **probabilities for the individual classes.** For example, if output one equals 0.4, output two equals 0.6, and the others are close to zero, then we can say that it's a 40% probability that the object in the picture is of class one and a 60% probability of class two. # Regression

  Let's now look at a **regression** problem where we want to** predict a numerical value rather than a class. **For example, we might try to predict the demand or price for an item.

In this case, we will use the **linear activation function.** **The linear activation is actually the same thing as no activation function at all.** **The output is simply the weighted sum.** This also means that the **output of this unit is unbounded and can be negative.**

 When we are solving a regression problem and are using the linear activation function, **we want to use the mean squared error as a loss function**.

We saw that depending on what problem we were trying to solve, we would pick different loss function. This is to facilitate the learning process, but the loss functions are not necessarily intuitive to interpret. Therefore, it's common to also compute other **metrics**. These metrics will not be part of the learning process itself, but **will give some intuition about the network behavior during learning.**

# Accuracy metric

 For example, if we're working on a **classification** problem, we typically want to compute the **accuracy metric.** This tells us **what percentage of the classifications are correct**.

For regression problems, accuracy is not that meaningful. It's very rare that the predictions will be exactly correct. Instead, what we want to compute is the **mean absolute error**, which measures the average prediction error during training.
