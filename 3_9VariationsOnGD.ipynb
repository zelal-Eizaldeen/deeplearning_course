# Variations on Gradient Descent

We will now go over some variations on gradient descent. In the previous videos, we **adjusted the weights** using **gradient descent and back-propagation**. In reality, we were implementing a special case of gradient descent, known as **stochastic gradient descent**.

Screenshot 2025-10-09 at 8.15.54â€¯AM.png

Formally, when we talk about **gradient descent, we would compute the partial derivatives of all training examples**. However, we did a single training example, computing the derivative for all training examples is also known as **batch gradient descent**.

That is, we would evaluate a **full batch before we do the weight adjustments.** While if **we just do a single example, we get an approximate value of the gradient**, and we call this **stochastic gradient descent**.

Screenshot 2025-10-09 at 8.16.33â€¯AM.png

There is a trade off here between accuracy and computation. If we compute the gradient for more training examples before we adjust, we will get more accurate adjustment. But if we do more computations before adjustment, we will slow down learning. A middle ground is best, we want to instead evaluate a small number of examples before we adjust the weights. This is known as **mini-batch gradient descent.**

Screenshot 2025-10-09 at 8.18.04â€¯AM.png

**Mini-batch gradient descent** works well on modern GPU hardware. A GPU can compute a full mini-batch in parallel as long as the mini-batch is of a reasonable size. Therefore, the time between weight adjustments is not affected by if we do a single example or a full mini-batch before adjusting the weights.

# Optimizers

Apart from the number of training examples between adjustments, there are other tweaks that can be done to the gradient descent algorithm to **enhance learning**.

In our **basic gradient descent algorithm**, we computed the gradient and adjusted the weight. We used a fixed step size, and then we looked at the next gradient. There are two key modifications that we can introduce.


1. The first one is **momentum**. This will consider not only the current gradient, but also the **values of previous gradients**.

# analogy
Think of a ball rolling down a hill. The ball's direction will be affected both of the slope in the current point, as well as the **momentum** that has built up based on previous slopes.

2. The second modification is that we can adjust the **step size** over time.

There are several ways of doing this. In a deep learning framework, we will use something called **optimizers** to choose between these different alternatives.
- If we want to produce just **momentum**, we can use a feature known as **Nesterov momentum**.
- If we want **adaptive step size**, we can use the AdaGrad or RMSProp optimizer.
- If we want to have both **momentum and adaptive step size**, we can use the Adam optimizer.

I recommend that you start with the **Adam optimizer** and then experiment with other optimizers if needed. In the code examples, you will see examples of using different optimizers.
