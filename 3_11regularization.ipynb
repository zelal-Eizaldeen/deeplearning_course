# Regularization

- We will now move on to a topic known as **regularization** techniques.

 We saw in previous examples that when we train the network, we want to train on the training set and test on the test set, and during this process, we want to monitor both the training error and the test error.
 - If the **training error** is not decreasing, the network architecture might be **unfit or insufficient capacity**. The solution to this would be to add **more layers or more neurons or use a different network architecture.**

- if the **training error decreases,** but the **test error plateaus or increases,** then we talk about **overfitting**. **This means that the network might memorize the training set, but it fails to generalize to the test set.**

# Regularization

Techniques used to address overfitting are known as **regularization techniques**.

The purpose of these techniques are **to reduce the gap between training and the test error.**

# 1. Early Stopping

In some cases, **both the training and test error will start decreasing to begin with, but then the test error starts to rise**. In such a case, we can use a technique known as **early stopping.** We will simply **stop training** at the point of **the smallest test error**.

However, it can sometimes be hard to determine where this point is because the error might statistically rise and fall over each epic.


The solution here will be **to periodically save the model weights,** and once we've **gotten to the point where we know where the minimum error was, we can reload that stored weights for that model.**

# 2.Weight decay

Another common regularization technique is **weight decay**. It's implemented by adding **a penalty term to the loss function**. Here is **L1 regularization**. We modify **the cross-entropy loss function by also adding a penalty term**. We will be **multiplying the absolute value** **of the weight **by a constant lambda **bold text**. **We add them together**.

For the **L2 regularization**, we will use the **squared weights rather than the absolute weights**. Since the learning algorithm strives to minimize the loss, this penalty term will imply that it also will try **to minimize the weights.** This will result in decreasing weights that solve only some of the examples, because these weights will incur a penalty for all other examples, but without much benefit. This is how weight decay will lead to better generalization.

#3. Dropout Regularization

Perhaps the most effective regularization techniques for neural network is called **dropout regularization**. Here we **randomly drop neurons during training**. We drop different neurons between each mini batch. According to the paper where this was introduced, this will force the units to work with random sets of units.
